```python
# 할 일 목록을 저장할 리스트
todo_list = []

# 할 일 목록을 표시하는 함수
def display_todo_list():
    """
    할 일 목록을 화면에 표시합니다.
    """
    if len(todo_list) == 0:                     # 리스트에 내용 없으면
        print("할 일 목록이 비어있습니다.")     # 목록 비었다고 출력
    else:
        for idx, item in enumerate(todo_list):  # enumerate으로 보여 줄때만 순서 부여, 순서는 데이터 저장 안됨
            print(f"{idx + 1}. {item}")         # 할일 리스트 존재 시 번호 부여 하고 출력


# 할 일을 추가하는 함수
def add_todo_item(item):
    """
    할 일 목록에 새로운 항목을 추가합니다.

    Args:
    item (str): 할 일 목록에 추가할 항목
    """
    todo_list.append(item)
    print(f"'{item}'이(가) 할 일 목록에 추가되었습니다.")

# 할 일을 삭제하는 함수
def delete_todo_item(index):
    """
    할 일 목록에서 특정 항목을 삭제합니다.

    Args:
    index (int): 삭제할 항목의 인덱스 (1부터 시작)
    """
    if 0 < index <= len(todo_list):
        item = todo_list.pop(index - 1)
        print(f"'{item}'이(가) 할 일 목록에서 삭제되었습니다.")
    else:
        print("잘못된 인덱스입니다.")

# 할 일을 모두 삭제하는 함수
def delete_todo_item_all():
    if len(todo_list) > 0:
        todo_list.clear()
        print("전부 삭제!")
    else:
        print("할일이 없습니다!")

# 사용자 입력 처리 함수
def handle_user_input():
    """
    사용자 입력을 처리합니다.
    """
    while True:
        print("\n할 일 목록 관리 프로그램")
        print("1. 할 일 표시하기")
        print("2. 할 일 추가하기")
        print("3. 할 일 삭제하기")
        print("4. 모든 일 삭제하기")
        print("5. 프로그램 종료하기")
        choice = input("원하는 작업을 선택하세요: ")

        if choice == '1':
            display_todo_list()
        elif choice == '2':
            item = input("추가할 할 일: ")
            add_todo_item(item)
        elif choice == '3':
            index = int(input("삭제할 할 일의 번호: "))
            delete_todo_item(index)
        elif choice == '4':
            delete_todo_item_all()
        elif choice == '5':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 시도해주세요.")




# 프로그램 실행
handle_user_input()
```
